```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
```

### Câu 1

```python
# Cho dữ liệu lưu trong tập tin "weather.csv"
weather = pd.read_csv('weather.csv')
weather.head()
```

```python
# Câu 1.1: Dùng Pandas vẽ biểu đồ dạng Box của 3 cột 'TMIN', 'TAVG', 'TMAX' và tiêu đề là 'Biểu đồ thời tiết'
# weather.info()
# weather.shape
weather[['TMIN', 'TAVG', 'TMAX']].plot.box(title='Biểu đồ thời tiết')
```

```python
# Câu 1.2: Thêm 1 cột dữ liệu 'TDIFF' là sự chênh lệch dữ liệu 'TMAX' - 'TMIN'
weather['TDIFF'] = weather['TMAX'] - weather['TMIN']
weather.head()
```

```python
# Câu 1.3: Dùng Pandas vẽ biểu đồ histogram của dữ liệu 'TDIFF' với 30 bins và giới hạn trục x là 0 và lưu biểu đồ vào tập tin "tdiff.png"
tdiff_weather = weather['TDIFF'].plot.hist(bins=30, xlim=0, title='Biểu đồ dữ liệu TDIFF')
fig = tdiff_weather.get_figure()
fig.savefig('tdiff.png')
```

```python
# Câu 1.4: Tạo một DataFrame 'bad_weather' mới từ các cột 'WT01' đến 'WT22'
bad_weather = weather.loc[:, 'WT01':'WT22']
bad_weather.head()
```

```python
# Câu 1.5: Tạo một cột mới 'bad_weather' trên DataFrame 'weather' bằng cách tính tổng theo cột trên DataFrame 'bad_weather'
weather['bad_weather'] = bad_weather.sum(axis=1)
weather.bad_weather
```

```python
# Câu 1.6: Điền các giá trị còn thiếu trên cột 'bad_weather' là 0 và chuyển kiểu dữ liệu của cột là int
weather.bad_weather = weather.bad_weather.fillna(0).astype('int')
weather.bad_weather
```

```python
# Câu 1.7: Dùng matplotlib vẽ và lưu (trong tập tin "bad_weather.png") biểu đồ histogram của cột dữ liệu 'bad_weather'. 
# Thiết lập số bins là 10, tiêu đề là 'Thời tiết', màu là 'green'
plt.hist(weather.bad_weather, bins=10, color='g')
plt.title('Thời tiết')
plt.savefig('bad_weather.png')
plt.show()
```
