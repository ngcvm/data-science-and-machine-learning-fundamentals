### Part 1: TreeMap
- Cho dữ liệu là danh sách ứng viên và số phiếu bầu trong cuộc bầu cử tổng thống Mỹ năm 2016.
- Vẽ 3 TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Virginia, Maryland và West Virginia


```python
# Import thư viện
import squarify
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
```

```python
# Cho dữ liệu: 
df = pd.DataFrame(
    {
        'Name': ['Hillary Clinton', 'Donald Trump', 'Others'],
        'Virginia': [1981473, 1769443, 233715],
        'Maryland': [1677928, 943169, 160349],
        'West Virginia': [188794, 489371, 36258],
    }
)

# In df
df
```

```python
# Vẽ TreeMap thể hiện tỷ lệ số phiếu bầu lần lượt cho ứng viên ở Virginia
phieu_bau_Vir = df['Virginia']
df.index.values
colors = [plt.cm.Spectral(i/float(len(phieu_bau_Vir))) for i in range(len(phieu_bau_Vir))]
squarify.plot(sizes=phieu_bau_Vir, label=df['Name'].values, color=colors)
plt.title('2016 Virginia President Election Results', fontdict={ 'color': 'blue' })
plt.show()
```

### Part 2: Waffle Chart
- Tính tổng số phiếu bầu của từng ứng viên ở cả 3 khu vực
- Vẽ Waffle Chart thể hiện tỷ lệ số phiếu bầu tổng cho từng ứng viên

```python
# Tính tổng số phiếu bầu của từng ứng viên ở cả 3 khu vực
df['Total'] = df.groupby('Name').transform('sum').sum().values
df
```

```python
# Vẽ Waffle Chart , lưu ý chia lại Tỷ lệ tổng giá trị là 1/100000
values = df.Total / 100000
fig = plt.figure(
    FigureClass=Waffle, 
    rows=df.shape[0], 
    values=values.values,
    labels=list(df.Name),
    legend={'loc': 'upper left', 'bbox_to_anchor': (1, 1)},
    figsize=(10, 5)
)
plt.title('2016 Presidential Election Results')
```
