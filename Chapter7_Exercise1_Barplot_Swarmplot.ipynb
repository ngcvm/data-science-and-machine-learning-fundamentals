```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
import collections
```

```python
# Câu 1: Cho dữ liệu:
x =  ["Excellent","Very Good","Good","Medium"]
y = [5.0,15.0,20.0,5.0]

# Hãy vẽ barplot
bp = sb.barplot(x, y)
# bp.title('Xếp loại')
plt.title('Xếp loại')
plt.show()
```

```python
# Câu 2: Cho dữ liệu
chars = "A A A A A A A A A A B B B B B B C C C C C C D D D D D D D D D D E E E E E E E F F F F G G G G G G G H H H H"
item_ocurrences = collections.Counter(chars.split(' '))
x = list(item_ocurrences.keys())
y = list(item_ocurrences.values())

# # Hãy vẽ barplot với mỗi cột có tên là ký tự, giá trị là số lần xuất hiện của ký tự đó trong chuỗi
cp= sb.color_palette('muted')
sb.set_style('darkgrid')
sb.barplot(x, y, palette=cp)
plt.title('Ký tự')
plt.show()
```

```python
# Câu 3: Biểu diễn dữ liệu data trên bằng swarmplot
df = pd.DataFrame(chars.split(' '))
df['value'] = 1
# df
cp= sb.color_palette('muted')
sb.swarmplot(x=df[0] , y=df['value'], data=df, palette=cp)
plt.title('Ký tự')
plt.ylabel('Giá trị')
plt.xlabel('Ký tự')
plt.show()
```
